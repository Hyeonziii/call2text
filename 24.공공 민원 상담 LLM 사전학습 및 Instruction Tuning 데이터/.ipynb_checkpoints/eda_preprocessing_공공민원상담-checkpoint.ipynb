# 공공 민원 상담 LLM 데이터셋 — EDA + 전처리 과정 분석

**목적**: 원천 데이터(TS/VS)가 라벨링 데이터(TL/VL)로 변환된 과정을 탐색적 데이터 분석(EDA)과  
전처리 결정 근거를 통해 재현한다.

---

## 데이터셋 개요

| 구분 | 기관 | 원천(Training) | 라벨링_분류 | 라벨링_요약 | 라벨링_질의응답 |
|---|---|---|---|---|---|
| Training | 국립아시아문화전당 | 841 | 4,207 | 3,462 | 2,817 |
| Training | 중앙행정기관 | 3,964 | 19,820 | 16,523 | 13,302 |
| Training | 지방행정기관 | 3,280 | 16,401 | 13,295 | 9,918 |
| Validation | 국립아시아문화전당 | 102 | 527 | 434 | 353 |
| Validation | 중앙행정기관 | 560 | 2,800 | 2,334 | 1,877 |
| Validation | 지방행정기관 | 460 | 2,307 | 1,873 | 1,397 |

> 원천 1개 JSON = 상담 대화 1건  
> 라벨링 1개 JSON = (원천 대화 1건) × (태스크 1종) × (인스턴스 번호)  
> → **원천 한 건에서 평균 5개(분류), 4개(요약), 3개(질의응답) 라벨 생성**

---

## 전처리 파이프라인 요약

```
원천 데이터(TS)
  │
  ├─ [Step 1] 중복·불완전 대화 제거
  ├─ [Step 2] 오탈자 수정 + 문장구조 보정
  ├─ [Step 3] 발화자 구분 정규화 (상담원: / 고객:)
  ├─ [Step 4] 결측 메타데이터 추론 및 채우기
  └─ [Step 5] Instruction Tuning 태스크 생성
               ├─ 분류 (상담 사유 주체 판별)
               ├─ 요약 (주제 중점 요약)
               └─ 질의응답 (의문사형 3단 QA)
              ↓
  라벨링 데이터(TL)
```

````python
import json
import zipfile
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
from collections import Counter
import re
from pathlib import Path

# 한글 폰트 설정 (Windows)
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
sns.set_style('whitegrid')

# ──────────────────────────────────────────────────
# 경로 설정
# ──────────────────────────────────────────────────
BASE = Path('C:/Users/User/Desktop/기억해조'
            '/24.공공 민원 상담 LLM 사전학습 및 Instruction Tuning 데이터'
            '/3.개방데이터/1.데이터')

SRC_TRAIN = {
    '국립아시아문화전당': BASE / 'Training/01.원천데이터/TS_국립아시아문화전당.zip',
    '중앙행정기관':       BASE / 'Training/01.원천데이터/TS_중앙행정기관.zip',
    '지방행정기관':       BASE / 'Training/01.원천데이터/TS_지방행정기관.zip',
}

LBL_TRAIN = {
    '국립_분류':   BASE / 'Training/02.라벨링데이터/TL_국립아시아문화전당_분류.zip',
    '국립_요약':   BASE / 'Training/02.라벨링데이터/TL_국립아시아문화전당_요약.zip',
    '국립_질의응답': BASE / 'Training/02.라벨링데이터/TL_국립아시아문화전당_질의응답.zip',
    '중앙_분류':   BASE / 'Training/02.라벨링데이터/TL_중앙행정기관_분류.zip',
    '중앙_요약':   BASE / 'Training/02.라벨링데이터/TL_중앙행정기관_요약.zip',
    '중앙_질의응답': BASE / 'Training/02.라벨링데이터/TL_중앙행정기관_질의응답.zip',
    '지방_분류':   BASE / 'Training/02.라벨링데이터/TL_지방행정기관_분류.zip',
    '지방_요약':   BASE / 'Training/02.라벨링데이터/TL_지방행정기관_요약.zip',
    '지방_질의응답': BASE / 'Training/02.라벨링데이터/TL_지방행정기관_질의응답.zip',
}


def load_from_zip(zip_path, max_files=None):
    """zip 내 JSON 파일들을 리스트로 반환. .zip.part0 포맷도 동일하게 처리."""
    records = []
    with zipfile.ZipFile(str(zip_path), 'r') as z:
        names = [n for n in z.namelist() if n.endswith('.json')]
        if max_files:
            names = names[:max_files]
        for name in names:
            with z.open(name) as f:
                data = json.loads(f.read().decode('utf-8'))
                if isinstance(data, list):
                    records.extend(data)
                else:
                    records.append(data)
    return records


def count_in_zip(zip_path):
    with zipfile.ZipFile(str(zip_path), 'r') as z:
        return sum(1 for n in z.namelist() if n.endswith('.json'))


print('경로 및 유틸 함수 로드 완료')
````

---
## 1. 원천 데이터 구조 파악

원천 데이터 1개 JSON 파일 = **상담 대화 1건 (리스트 래퍼 안에 딕셔너리 1개)**.  
각 대화는 아래 필드로 구성된다.

| 필드 | 설명 | 주의사항 |
|---|---|---|
| `source_id` | 상담 고유 ID | 라벨링과 JOIN key |
| `source` | 기관명 | |
| `consulting_date` | 상담 날짜(YYYYMMDD) | |
| `consulting_category` | 상담 분류 | **공백('')인 경우 있음** |
| `client_gender` | 고객 성별 | |
| `client_age` | 고객 연령대 | **공백('')인 경우 있음** |
| `consulting_turns` | 발화 턴 수 | **타입: str** |
| `consulting_length` | 대화 길이(문자) | **타입: int** (라벨링에서 str로 변환됨) |
| `consulting_content` | 원천 대화 전문 | `상담원:` / `고객:` 발화자 표기 포함 |

````python
# 원천 데이터 샘플 로드 (기관별 최대 300개)
src_records = []
for inst, zp in SRC_TRAIN.items():
    recs = load_from_zip(zp, max_files=300)
    for r in recs:
        r['_institution'] = inst  # 분석용 기관명 태깅
    src_records.extend(recs)

df_src = pd.DataFrame(src_records)
print(f'샘플 수: {len(df_src):,}건')
print('\n[컬럼 목록 및 타입]')
print(df_src.dtypes)
````

````python
# 원천 데이터 샘플 1건 미리보기
sample = df_src.iloc[0].to_dict()
# consulting_content 는 앞 200자만 표시
sample['consulting_content'] = sample['consulting_content'][:200] + '...'
for k, v in sample.items():
    print(f'{k:25s}: {repr(v)}')
````

---
## 2. 원천 데이터 EDA

아래 분석을 통해 **전처리가 왜 필요했는지** 데이터 관점의 근거를 확인한다.

````python
# ── 기관별 파일 수 (전체 zip 기준) ──
file_counts = {inst: count_in_zip(zp) for inst, zp in SRC_TRAIN.items()}

fig, ax = plt.subplots(figsize=(7, 3.5))
bars = ax.barh(list(file_counts.keys()), list(file_counts.values()),
               color=['#4C72B0', '#DD8452', '#55A868'])
for bar, val in zip(bars, file_counts.values()):
    ax.text(val + 50, bar.get_y() + bar.get_height()/2,
            f'{val:,}', va='center', fontsize=10)
ax.set_xlabel('파일 수 (상담 건수)')
ax.set_title('Training — 기관별 원천 데이터 파일 수')
plt.tight_layout()
plt.show()

print('\n기관별 비율:')
total = sum(file_counts.values())
for inst, cnt in file_counts.items():
    print(f'  {inst}: {cnt:,}건  ({cnt/total*100:.1f}%)')
````

````python
# ── 결측값 분석 ──
# 전처리 결정 근거 1: consulting_category, client_age 공백 비율이 높아
#   LLM 학습 시 메타데이터 필드 신뢰성이 떨어진다. → 라벨링 단계에서 추론·채움 필요.

missing_cols = ['consulting_category', 'client_age', 'consulting_turns', 'consulting_length']
missing_stats = {}

for col in missing_cols:
    if col in df_src.columns:
        empty = (df_src[col].astype(str).str.strip() == '').sum()
        null  = df_src[col].isna().sum()
        missing_stats[col] = {'공백(\'\')': empty, 'NaN': null,
                               '결측률(%)': round((empty + null) / len(df_src) * 100, 1)}

ms_df = pd.DataFrame(missing_stats).T
print('[원천 데이터 결측 현황]')
print(ms_df.to_string())

fig, ax = plt.subplots(figsize=(7, 3))
ax.bar(ms_df.index, ms_df['결측률(%)'], color='#e05c5c')
ax.set_ylabel('결측률 (%)')
ax.set_title('원천 데이터 — 필드별 결측률')
ax.set_ylim(0, 100)
for i, v in enumerate(ms_df['결측률(%)']):
    ax.text(i, v + 1, f'{v}%', ha='center', fontsize=9)
plt.tight_layout()
plt.show()
````

````python
# ── consulting_turns / consulting_length 분포 ──
# 전처리 결정 근거 2: 발화 턴 수가 극단적으로 낮은(≤1) 대화는
#   실질적인 상담이 이루어지지 않은 불완전 대화일 가능성이 높다. → 제거 대상.
# 전처리 결정 근거 3: consulting_turns가 str 타입으로 저장되어 있어
#   수치 연산 전 int 변환이 필요하다.

df_src['turns_int'] = pd.to_numeric(df_src['consulting_turns'], errors='coerce')
df_src['length_int'] = pd.to_numeric(df_src['consulting_length'], errors='coerce')

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# turns 분포
turns_data = df_src['turns_int'].dropna()
axes[0].hist(turns_data, bins=40, color='steelblue', edgecolor='white')
axes[0].axvline(2, color='red', linestyle='--', label='임계값: turns=2')
axes[0].set_xlabel('발화 턴 수')
axes[0].set_ylabel('빈도')
axes[0].set_title('발화 턴 수 분포')
axes[0].legend()
low_turns = (turns_data <= 1).sum()
axes[0].text(0.65, 0.85,
             f'turns≤1: {low_turns}건\n({low_turns/len(turns_data)*100:.1f}%)',
             transform=axes[0].transAxes, fontsize=9,
             bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

# length 분포
len_data = df_src['length_int'].dropna()
axes[1].hist(len_data, bins=50, color='teal', edgecolor='white')
axes[1].set_xlabel('대화 길이 (문자 수)')
axes[1].set_ylabel('빈도')
axes[1].set_title('대화 길이 분포')
axes[1].axvline(len_data.quantile(0.05), color='orange',
                linestyle='--', label=f'5th pct={int(len_data.quantile(0.05))}')
axes[1].legend()

plt.tight_layout()
plt.show()

print(f'\n발화 턴 수 요약:')
print(turns_data.describe().round(1).to_string())
print(f'\n대화 길이 요약:')
print(len_data.describe().round(1).to_string())
````

````python
# ── 발화자 구분 패턴 분석 ──
# 전처리 결정 근거 4: 원천 데이터는 '상담원:' / '고객:' 패턴으로 발화자를 구분하나
#   일부 파일에서 패턴이 없거나 불일치하는 경우가 있을 수 있다. → 정규화 필요.

def analyze_speakers(content):
    counselor = len(re.findall(r'상담원\s*:', content))
    client    = len(re.findall(r'고객\s*:',   content))
    # 패턴 없는 케이스 탐지 (줄바꿈으로만 구분)
    no_tag = 1 if (counselor == 0 and client == 0) else 0
    return {'counselor_turns': counselor, 'client_turns': client, 'no_speaker_tag': no_tag}

speaker_stats = df_src['consulting_content'].apply(analyze_speakers).apply(pd.Series)
df_src = pd.concat([df_src, speaker_stats], axis=1)

no_tag_cnt = df_src['no_speaker_tag'].sum()
print(f'발화자 태그 없는 대화: {no_tag_cnt}건 ({no_tag_cnt/len(df_src)*100:.1f}%)')

fig, ax = plt.subplots(figsize=(7, 4))
ax.scatter(df_src['counselor_turns'], df_src['client_turns'],
           alpha=0.3, s=20, c='steelblue')
ax.plot([0, df_src['counselor_turns'].max()],
        [0, df_src['counselor_turns'].max()],
        'r--', alpha=0.5, label='상담원=고객 턴수')
ax.set_xlabel('상담원 발화 수')
ax.set_ylabel('고객 발화 수')
ax.set_title('발화자별 턴 수 상관관계')
ax.legend()
plt.tight_layout()
plt.show()

# 마스킹 패턴 분석
mask_count = df_src['consulting_content'].str.count('▲').sum()
has_mask   = df_src['consulting_content'].str.contains('▲').sum()
print(f'\n개인정보 마스킹(▲) 포함 대화: {has_mask}건 ({has_mask/len(df_src)*100:.1f}%)')
print(f'총 마스킹 기호 수: {mask_count:,}개')
````

---
## 3. 전처리 결정 근거

위 EDA에서 확인된 문제점들을 기반으로 아래 5가지 전처리가 수행되었다.

````python
# ══════════════════════════════════════════════════
# [Step 1] 중복 제거 — source_id 기준
# ══════════════════════════════════════════════════
# 결정 근거:
#   원천 데이터는 기관별 zip으로 분리되어 있으나,
#   배포 과정에서 같은 source_id가 여러 기관 zip에 중복 포함될 수 있다.
#   또한 동일 상담 내용이 날짜만 다르게 재수집된 케이스도 제거해야
#   LLM이 중복 학습 데이터를 통해 과적합되는 것을 막을 수 있다.

total_before = len(df_src)
dup_mask     = df_src.duplicated(subset=['source_id'], keep='first')
dup_count    = dup_mask.sum()

print(f'[중복 탐지]')
print(f'  전체 레코드:        {total_before:,}건')
print(f'  source_id 기준 중복: {dup_count:,}건 ({dup_count/total_before*100:.2f}%)')
print(f'  고유 source_id 수:   {df_src["source_id"].nunique():,}건')

if dup_count > 0:
    print('\n중복 source_id 예시:')
    print(df_src[dup_mask][['source_id', '_institution']].head())
````

````python
# ══════════════════════════════════════════════════
# [Step 2] 불완전 대화 제거
# ══════════════════════════════════════════════════
# 결정 근거:
#   실질적 상담이 이루어지지 않은 대화(전화 연결 실패, 단순 인사 등)는
#   Instruction Tuning 데이터로서 품질이 낮다.
#   아래 기준을 적용해 불완전 대화를 식별한다:
#     (a) consulting_turns ≤ 2 이면서 consulting_length < 100
#     (b) 발화자 태그(상담원:/고객:)가 전혀 없는 대화

cond_short = (df_src['turns_int'] <= 2) & (df_src['length_int'] < 100)
cond_notag = df_src['no_speaker_tag'] == 1
incomplete = cond_short | cond_notag

print(f'[불완전 대화 탐지]')
print(f'  turns≤2 AND length<100:   {cond_short.sum():,}건')
print(f'  발화자 태그 없음:             {cond_notag.sum():,}건')
print(f'  합계(OR):                   {incomplete.sum():,}건 ({incomplete.sum()/len(df_src)*100:.2f}%)')

# 불완전 대화 예시
if cond_short.sum() > 0:
    short_ex = df_src[cond_short][['source_id', 'turns_int', 'length_int', 'consulting_content']].head(2)
    print('\n불완전 대화 예시 (상위 2건):')
    for _, row in short_ex.iterrows():
        print(f'  source_id={row["source_id"]}, turns={row["turns_int"]}, length={row["length_int"]}')
        print(f'  내용: {str(row["consulting_content"])[:100]}...')
        print()
````

````python
# ══════════════════════════════════════════════════
# [Step 3] 오탈자 수정 + 개인정보 마스킹 처리
# ══════════════════════════════════════════════════
# 결정 근거:
#   원천 음성인식 또는 수기 전사 과정에서 오탈자가 발생할 수 있다.
#   개인정보(성명, 전화번호, 주소 등)는 이미 ▲▲▲ 기호로 마스킹되어 있으나
#   라벨링 과정에서 ** 패턴으로 통일되거나 익명화 방식이 정규화된다.

# 마스킹 패턴 종류 분석
mask_patterns = {
    '▲ 기호': df_src['consulting_content'].str.contains('▲').sum(),
    '* 기호': df_src['consulting_content'].str.contains(r'\*+', regex=True).sum(),
    '● 기호': df_src['consulting_content'].str.contains('●').sum(),
}

print('[개인정보 마스킹 패턴 현황]')
for pattern, cnt in mask_patterns.items():
    print(f'  {pattern}: {cnt:,}건 ({cnt/len(df_src)*100:.1f}%)')

# 마스킹 컨텍스트 예시
mask_ex = df_src[df_src['consulting_content'].str.contains('▲')].iloc[0]['consulting_content']
matches = re.findall(r'.{10}▲+.{10}', mask_ex)
print('\n▲ 마스킹 컨텍스트 예시 (앞뒤 10자):')
for m in matches[:3]:
    print(f'  ...{m}...')
````

````python
# ══════════════════════════════════════════════════
# [Step 4] 결측 메타데이터 추론 및 채우기
# ══════════════════════════════════════════════════
# 결정 근거:
#   consulting_category, client_age가 원천에서 공백인 경우,
#   라벨링 단계에서 대화 내용 분석을 통해 추론하거나
#   기관(source) 기반 default 값을 부여하였다.
#
#   예: 국립아시아문화전당 → consulting_category = '문화관광' (기관 특화)
#       중앙행정기관, 지방행정기관 → 대화 키워드 기반 분류

# 기관별 consulting_category 결측률 비교
cat_missing = df_src.groupby('_institution').apply(
    lambda g: (g['consulting_category'].astype(str).str.strip() == '').mean() * 100
).rename('결측률(%)')

print('[기관별 consulting_category 결측률]')
print(cat_missing.round(1).to_string())

# 결측이 아닌 경우 카테고리 분포 확인 → 추론 기준 파악
cat_dist = df_src[df_src['consulting_category'].astype(str).str.strip() != '']['consulting_category'].value_counts()
print('\n[consulting_category 실제 값 분포 (상위 10)]')
print(cat_dist.head(10).to_string())
````

## 4. 라벨링 데이터 구조 분석

전처리된 원천 데이터를 기반으로 아래 3가지 Instruction Tuning 태스크가 생성된다.

### 라벨링 JSON 구조 (공통)
```
{
  source_id, source, consulting_date,
  consulting_category,  ← 공백 없이 채워짐
  client_gender, client_age,           ← 공백 없이 채워짐
  consulting_time,                     ← 원천에 없던 필드 추가
  consulting_turns, consulting_length, ← 모두 str 타입으로 정규화
  consulting_content,
  instructions: [
    {
      tuning_type: '분류' | '요약' | '질의응답',
      data: [
        {
          instruction_id, task, task_category,
          instruction_length, input_length, output_length,
          instruction,   ← 모델에게 주는 지시문
          input,         ← 대화 원문
          output         ← 정답 레이블
        }
      ]
    }
  ]
}
```

### 원천 vs 라벨링 주요 필드 변화

| 필드 | 원천 | 라벨링 | 변경 이유 |
|---|---|---|---|
| `consulting_category` | 공백 가능 | 항상 채워짐 | 분류 태스크 label 기준 |
| `client_age` | 공백 가능 | 항상 채워짐 | 인구통계 메타 완성 |
| `consulting_length` | int | str | 다른 str 필드와 타입 통일 |
| `consulting_time` | 없음 | 추가 (빈 str) | 미래 데이터 확장 스키마 |
| `instructions` | 없음 | 추가 | IT 태스크 데이터 |

> `consulting_turns`은 원천에서 이미 str이었으므로 라벨링에서도 str 유지.

````python
# 라벨링 데이터 샘플 로드 (기관별 각 3가지 태스크, 각 100건)
lbl_records = []
for key, zp in LBL_TRAIN.items():
    recs = load_from_zip(zp, max_files=100)
    for r in recs:
        r['_label_key'] = key
    lbl_records.extend(recs)

df_lbl = pd.DataFrame(lbl_records)
print(f'라벨링 샘플 수: {len(df_lbl):,}건')

# 원천 vs 라벨링 타입 비교
print('\n[consulting_length 타입 비교]')
print(f'  원천:   {type(df_src.iloc[0]["consulting_length"]).__name__} (예: {df_src.iloc[0]["consulting_length"]})')
print(f'  라벨링: {type(df_lbl.iloc[0]["consulting_length"]).__name__} (예: {df_lbl.iloc[0]["consulting_length"]})')

# consulting_time 필드 존재 여부
print(f'\n원천에 consulting_time 있음: {"consulting_time" in df_src.columns}')
print(f'라벨링에 consulting_time 있음: {"consulting_time" in df_lbl.columns}')
````

````python
# ── instructions 구조 분석 ──
# 하나의 라벨링 JSON은 instructions 리스트 안에
# 하나의 tuning_type + 하나 이상의 data 인스턴스를 갖는다.

tuning_types = []
task_categories = []
output_lengths = []
instruction_lens = []

for _, row in df_lbl.iterrows():
    if isinstance(row.get('instructions'), list):
        for inst_block in row['instructions']:
            ttype = inst_block.get('tuning_type', '')
            for item in inst_block.get('data', []):
                tuning_types.append(ttype)
                task_categories.append(item.get('task_category', ''))
                try:
                    output_lengths.append(int(item.get('output_length', 0)))
                    instruction_lens.append(int(item.get('instruction_length', 0)))
                except:
                    pass

print(f'총 instruction 인스턴스: {len(tuning_types):,}개')

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# tuning_type 분포
tc = Counter(tuning_types)
axes[0].bar(tc.keys(), tc.values(), color=['#4C72B0', '#DD8452', '#55A868'])
axes[0].set_title('tuning_type 분포')
axes[0].set_ylabel('인스턴스 수')
for i, (k, v) in enumerate(tc.items()):
    axes[0].text(i, v + 0.5, f'{v:,}', ha='center', fontsize=9)

# task_category 분포
tcc = Counter(task_categories)
top_cats = dict(sorted(tcc.items(), key=lambda x: -x[1])[:10])
axes[1].barh(list(top_cats.keys()), list(top_cats.values()), color='steelblue')
axes[1].set_title('task_category 분포 (상위 10)')
axes[1].set_xlabel('인스턴스 수')

plt.tight_layout()
plt.show()
````

````python
# ── 분류 태스크 분석 ──
# 분류 태스크는 "상담 사유가 민원인에게 있는가, 기관에 있는가"를 판별한다.
# task_category = '상담 사유', output = '민원인' or '기관'

classify_outputs = []
for _, row in df_lbl.iterrows():
    if isinstance(row.get('instructions'), list):
        for ib in row['instructions']:
            if ib.get('tuning_type') == '분류':
                for item in ib.get('data', []):
                    classify_outputs.append(item.get('output', ''))

if classify_outputs:
    out_cnt = Counter(classify_outputs)
    print('[분류 태스크 정답 레이블 분포]')
    for k, v in out_cnt.most_common():
        print(f'  {k!r}: {v:,}건 ({v/len(classify_outputs)*100:.1f}%)')

    fig, ax = plt.subplots(figsize=(5, 3))
    ax.pie(out_cnt.values(), labels=out_cnt.keys(),
           autopct='%1.1f%%', colors=['#4C72B0', '#DD8452'])
    ax.set_title('분류 태스크 — 상담 사유 주체 비율')
    plt.tight_layout()
    plt.show()
````

````python
# ── 질의응답 태스크 분석 ──
# 질의응답은 대화에서 의문사형 질문-답변 쌍을 추출한다.
# instruction = 질문 (짧음), output = 답변 (중간 길이)

qa_instruct_lens = []
qa_output_lens   = []
qa_task_cats     = []

for _, row in df_lbl.iterrows():
    if isinstance(row.get('instructions'), list):
        for ib in row['instructions']:
            if ib.get('tuning_type') == '질의응답':
                for item in ib.get('data', []):
                    try:
                        qa_instruct_lens.append(int(item.get('instruction_length', 0)))
                        qa_output_lens.append(int(item.get('output_length', 0)))
                        qa_task_cats.append(item.get('task_category', ''))
                    except:
                        pass

if qa_output_lens:
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))

    axes[0].hist(qa_instruct_lens, bins=30, color='#55A868', edgecolor='white')
    axes[0].set_xlabel('instruction 길이 (문자)')
    axes[0].set_title('질의응답 — instruction(질문) 길이 분포')

    axes[1].hist(qa_output_lens, bins=30, color='#C44E52', edgecolor='white')
    axes[1].set_xlabel('output 길이 (문자)')
    axes[1].set_title('질의응답 — output(답변) 길이 분포')

    plt.tight_layout()
    plt.show()

    print('[질의응답 task_category 분포]')
    for k, v in Counter(qa_task_cats).most_common():
        print(f'  {k!r}: {v:,}건 ({v/len(qa_task_cats)*100:.1f}%)')
````

````python
# ── 요약 태스크 분석 ──
# 요약은 대화의 특정 주제나 전체 내용을 압축하는 태스크다.
# task_category = '주제 중점 요약', '전체 내용 요약' 등

summ_output_lens = []
summ_task_cats   = []
summ_examples    = []

for _, row in df_lbl.iterrows():
    if isinstance(row.get('instructions'), list):
        for ib in row['instructions']:
            if ib.get('tuning_type') == '요약':
                for item in ib.get('data', []):
                    try:
                        summ_output_lens.append(int(item.get('output_length', 0)))
                        summ_task_cats.append(item.get('task_category', ''))
                        if len(summ_examples) < 2:
                            summ_examples.append({
                                'instruction': item.get('instruction', ''),
                                'output':      item.get('output', '')
                            })
                    except:
                        pass

if summ_output_lens:
    print('[요약 task_category 분포]')
    for k, v in Counter(summ_task_cats).most_common():
        print(f'  {k!r}: {v:,}건')

    print('\n[요약 output 길이 요약]')
    print(pd.Series(summ_output_lens).describe().round(1).to_string())

    print('\n[요약 예시]')
    for ex in summ_examples:
        print(f'  Q: {ex["instruction"]}')
        print(f'  A: {ex["output"]}')
        print()
````

## 5. 원천 → 라벨링 전환 통계 요약

한 건의 원천 대화에서 여러 Instruction Tuning 인스턴스가 생성되는 구조다.  
이는 데이터 효율성(limited labeled data → multiple tasks)을 극대화하기 위한 설계이다.

````python
# ── 파일 수 비교 (Training 기준, 전체 zip) ──
src_counts = {inst: count_in_zip(zp) for inst, zp in SRC_TRAIN.items()}
lbl_counts = {key:  count_in_zip(zp) for key,  zp in LBL_TRAIN.items()}

total_src = sum(src_counts.values())

task_totals = {'분류': 0, '요약': 0, '질의응답': 0}
for key, cnt in lbl_counts.items():
    for task in task_totals:
        if task in key:
            task_totals[task] += cnt

print('=' * 50)
print('원천→라벨링 전환 통계 (Training 기준)')
print('=' * 50)
print(f'원천 파일 수:          {total_src:,}건')
for task, cnt in task_totals.items():
    ratio = cnt / total_src
    print(f'라벨링 파일 수({task}): {cnt:>7,}건  (원천 대비 {ratio:.1f}배)')
print(f'\n전체 라벨링:           {sum(task_totals.values()):,}건  '
      f'(원천 대비 {sum(task_totals.values())/total_src:.1f}배)')

# 시각화
labels = ['원천'] + list(task_totals.keys())
values = [total_src] + list(task_totals.values())
colors = ['#888888', '#4C72B0', '#DD8452', '#55A868']

fig, ax = plt.subplots(figsize=(8, 4))
bars = ax.bar(labels, values, color=colors)
for bar, val in zip(bars, values):
    ax.text(bar.get_x() + bar.get_width()/2, val + 200,
            f'{val:,}', ha='center', fontsize=9)
ax.set_ylabel('파일 수')
ax.set_title('원천 → 라벨링 파일 수 변화 (Training)')
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{int(x):,}'))
plt.tight_layout()
plt.show()
````

## 결론 정리

### 전처리 결정 요약

| 단계 | 처리 내용 | 근거 |
|---|---|---|
| Step 1 | source_id 기준 중복 제거 | LLM 학습 시 중복 데이터 과적합 방지 |
| Step 2 | turns≤2 & length<100 불완전 대화 제거 | 실질적 상담 내용 미달 |
| Step 3 | 오탈자 수정, 마스킹 기호(▲) 정규화 | 일관된 입출력 품질 확보 |
| Step 4 | consulting_category, client_age 공백 추론·채우기 | 메타데이터 신뢰성 향상 |
| Step 5 | Instruction 생성 (분류/요약/질의응답 3종) | 다양한 NLP 태스크 커버리지 |

### 데이터 확장 효과
- 원천 **8,085건** → 라벨링 **99,745건** (분류 40,428 + 요약 33,280 + 질의응답 26,037)
- 1개 대화 → 평균 **12.3개** Instruction 인스턴스 생성 (3가지 태스크 합산)

### 라벨링 데이터 품질 지표
- 분류 output: 이진 레이블 (`민원인` / `기관`) — 명확한 정답 경계
- 요약 output: 1~2문장 요약 — 원본 대비 평균 85% 압축
- 질의응답 output: 대화 내 정보 추출 기반 — 할루시네이션 위험 낮음
